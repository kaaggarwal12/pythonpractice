```python
#You have a list of words, and you want to process them according to the following rules:

#Skip any word that starts with the letter 's' .
#If the word is exactly "stop", do nothing and move to the next word .
#If the word is "exit", stop processing further words .
#For all other words, print them in uppercase.
```

```python
words = ["apple", "strawberry", "banana", "stop", "cherry", "exit", "date", "grape", "stoplight"]
```

```python
for x in words:
    if x.startswith('s'):
        continue
    elif x == "stop":
        pass
    elif x == "exit":
        break
    else:
        print(x.upper())
     
```

```python
#Imagine we have a list of numbers, and we want to process them in the following way:

#Skip any negative numbers.
#If the number is zero, we simply want to do nothing .
#If the number is greater than 10, we want to stop processing further numbers .
#For all other numbers, we want to print them.
```

```python
numbers = [3, -1, 0, 8, 12, 7, -5, 6]
```

```python
for x in numbers:
    if x < 0:
        continue
    elif x == 0:
        pass
    elif x > 10:
        break
    else:
        print(x)
```

```python
#You have a list of transactions, where each transaction is represented as a dictionary with keys: "type" (either "credit" or "debit") and "amount".
#You need to process these transactions according to the following rules:

#Skip any transaction where the amount is negative (use continue).
#If the transaction type is "debit" and the amount is exactly 0, do nothing and move to the next transaction (use pass).
#If the transaction type is "credit" and the amount exceeds $1000, stop processing further transactions (use break).
#For all other transactions, print the transaction details.
```

```python
transactions = [
    {"type": "credit", "amount": 500},
    {"type": "debit", "amount": -200},
    {"type": "debit", "amount": 0},
    {"type": "credit", "amount": 1200},
    {"type": "debit", "amount": 300},
    {"type": "credit", "amount": 100},
    {"type": "credit", "amount": 800},
    {"type": "debit", "amount": 50},
]
```

```python
for x in transactions:
    if x['amount'] < 0:
        continue
    elif x["type"] == "debit" and x["amount"] == 0:
        pass
    elif x["type"] == "credit" and x['amount'] > 1000:
        break
    else:
        print(x)
```

```python
transactions = [
    {"type": "credit", "amount": 500},
    {"type": "debit", "amount": -200},
    {"type": "debit", "amount": 0},
    {"type": "credit", "amount": 1200},
    {"type": "debit", "amount": 300},
    {"type": "credit", "amount": 100},
    {"type": "credit", "amount": 800},
    {"type": "debit", "amount": 50},
]

for transaction in transactions:
    if transaction["amount"] < 0:
        continue  # Skip transactions with negative amounts
    elif transaction["type"] == "debit" and transaction["amount"] == 0:
        pass  # Do nothing for debit transactions with amount 0
    elif transaction["type"] == "credit" and transaction["amount"] > 1000:
        break  # Stop processing if credit amount exceeds 1000
    else:
        print(transaction)  # Print other transactions
```
